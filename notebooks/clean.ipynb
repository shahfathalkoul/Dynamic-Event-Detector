```python
data['full_text'] = data['headline'] + ' ' + data['short_description'].fillna('')
data['full_text'].head(3)
```

```python
data_sample = data.sample(n=10000, random_state=42).reset_index(drop=True)
data_sample.shape
```

```python
#Clean function 
import re
import nltk
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z\s]', '', text)
    words = [lemmatizer.lemmatize(w)
             for w in text.split()
             if w not in stop_words and len(w) > 2]
    return ' '.join(words)
```

```python
data_sample['clean_text'] = data_sample['full_text'].apply(clean_text)
data_sample[['full_text', 'clean_text']].head(3)
```

```python
import os
os.makedirs('../data', exist_ok=True)
data_sample.to_csv('../data/clean_data.csv', index=False)
data_sample.shape
```
